In this notebook, we example data from number of unique sums (num_hands) that can be formed
from a randomly chosen 8-card hand, and the corresponding number of unique digits in that hand.
The results show that the number of unique sums increase with the number of
unique digits in that hand.  Also, the distribution for num_hands is weighted to the left,
which means one is more likely to have an 8-card hand with relatively few possible sums
that can be formed.

```python
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("num_hands_with_digits_10-card.csv")
df["num_sums"] = df["num_hands"]
df.drop("num_hands",axis=1,inplace=True)
```

```python
df.head()
```

```python
sns.histplot(data=df,x="num_sums",bins=15)
```

```python
sns.boxplot(data=df, x="num_unique_digits", y="num_sums")
```

```python
sns.histplot(data=df, x="num_unique_digits")
```
